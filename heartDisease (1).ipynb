```python
df.age.replace(0, np.nan, inplace=True)

df.trestbps.replace(0, np.mean(df.trestbps), inplace=True)
df.chol.replace(0, np.mean(df.chol), inplace=True)
df = df.dropna()
from sklearn.utils import shuffle
df = shuffle(df)
df.head(20)
```

```python
import pandas as pd
from sklearn import preprocessing


min_max_scaler = preprocessing.MinMaxScaler()
columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
df[columns_to_scale] = min_max_scaler.fit_transform(df[columns_to_scale])
```

```python
import keras
from keras.layers import Dense
from keras.models import Sequential
from keras.utils import to_categorical
```

```python
predictors = df.drop('target', axis=1).as_matrix()
targett = to_categorical(df.target)
model = Sequential()
n_cols = predictors.shape[1]
```

```python
model.add(Dense(100, activation='relu', input_shape=(n_cols,)))
model.add(Dense(100, activation='relu'))

model.add(Dense(2, activation='softmax'))
```

```python
model.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(predictors, targett, validation_split=0.25, epochs=37)
```

```python
pred = model.predict([[[0.645833,0,0,0.528302,0.301370,0,0,0.656489,0,0.419355,1,2,3]]])
print(pred)
print(np.argmax(pred))
```
